```

from sympy import symbols, I, simplify, pi, E

# Define las variables simbólicas
a1, a2, b1, b2 = symbols("a1 a2 b1 b2", real=True)

# Definir las variables complejas
a = a1 + I * a2
b = b1 + I * b2

# Expresar la división
resultado = a / b * b

# Simplificar la expresión
resultado_simplificado = simplify(resultado)

# Imprimir el resultado

print(expand(a/b*b))
print(expand(a/(b*b)))
print(expand((a/b)*b))




w =  (pi + I)/E
print('w =', w)

w_numeric = N(w)
print('w =', w_numeric)
```
